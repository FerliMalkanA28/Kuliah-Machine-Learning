## Library

```python
import pandas as pd
import numpy as np
import sklearn

from sklearn.cluster import AgglomerativeClustering
from sklearn import datasets
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

## Latihan 1

### Dataset

```python
# buat daftar tabular seperti di bawah ini
df = pd.DataFrame(
  {
    "ID Buah": pd.Series([1, 2, 3, 4, 5]),
    "Diameter": pd.Series([7.0, 7.1, 6.5, 6.6, 6.8]),
    "Berat" :pd.Series([165, 170, 180, 195, 200]),
  }
)
df
```

### Preprocessing

```python
# pemilihan feature
X = df.drop("ID Buah", axis=1)
X
```

### Modeling
Menggunakan algoritma agglomerative clustering

```python
clustering = AgglomerativeClustering().fit(X)
# hasil clustering
df["kluster"] = clustering.labels_
df
```

## Latihan 2

### Dataset

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/student_score_pass.csv")
df2
```

*Link data https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/student_score_pass.csv

### Preprocessing

```python
# pemilihan feature
df_training = df2.drop(['npm', 'pass_test'], axis=1)
# hapus NaN values
df_training = df_training.dropna()
df_training
```

### Modeling

```python
model = AgglomerativeClustering().fit(df_training)
# hasil clustering
df_training["kluster"] = model.labels_
df_training                                
```

# Tugas
Evaluasi model algoritma agglomerative clustering dari data latihan 2 dengan mencari nilai akurasinya. Hint : splitting data, modeling, accuracy score. <br> <br>
Nilai akurasi: 0.5079365079365079

```python
X = df_training.drop("kluster", axis=1)
y = df_training["kluster"]
```

### Preprocessing

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X)
```

```python
scaled_data = scaler.transform(X)
```

```python
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
```

```python
scaled_data.head().T
```

### Spitting Data

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=50)
```

### Modeling

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

### Accuracy Score

```python
model.score(X_test, y_test)
```
