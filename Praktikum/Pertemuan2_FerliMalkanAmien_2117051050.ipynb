```python
import pandas as pd
import numpy as np
# import apriori from library mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

## Dataset
*   Membuat dataset transaksi sebagai dataframe
*   Cek dataset transaksi

```python
# membuat data tabular yang terdiri dari data transaksi
df = pd.DataFrame(
    {
        "ID Transaksi" : pd.Series([1, 2, 3, 4, 5]),
        "Item" : pd.Series([["Susu", "Roti", "Biskuit", "Mentega"], \
                            ["Susu", "Biskuit", "Gula", "Kopi"], \
                            ["Gula", "Kopi", "Susu"], ["Kopi", "Biskuit", "Susu", "Gula"], \
                            ["Mentega", "Roti", "Kopi"]])
    }
)
```

```python
df.info()
```

```python
df
```

## Encoding data transaksi

```python
df = df.explode('Item')
df
```

```python
basket = (df.groupby(['ID Transaksi', 'Item'])['Item'].count()\
          .unstack().reset_index().fillna(0)\
          .set_index('ID Transaksi'))

basket.head()
```

## Algoritma Apriori
Eksekusi algoritma apriori menggunakan library mlextend dengan fungsi apriori

### Apriori Tahap 1
Eliminasi itemset berdasarkan minimum support

```python
# set minimum support 60%
frequent_itemsets = apriori(basket, min_support=0.6, use_colnames=True)
# print(list(frequent_itemsets))
frequent_itemsets
```

### Apriori Tahap 2
Perhitungan confidence

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules[["antecedents", "consequents", "support", "confidence"]]
```

## 2. Buat penyelesaian apriori menggunakan fungsi manual, atau fungsi yang dibuat sendiri tidak menggunakan library (bonus)

```python
def load_dataset():
    # Dataset dari soal sebelumnya
    dataset = [
        {'bread', 'milk', 'cheese'},
        {'bread', 'milk', 'diapers', 'beer'},
        {'bread', 'diapers', 'beer', 'eggs'},
        {'milk', 'diapers', 'beer', 'cheese'},
        {'bread', 'milk', 'diapers', 'beer', 'cheese'},
    ]
    return dataset

def create_itemsets(dataset):
    # Buat sebuah list dari semua item unik di dalam dataset
    items = set()
    for transaction in dataset:
        for item in transaction:
            items.add(frozenset([item]))
    return items

def calculate_support(dataset, itemsets, min_support):
    # Hitung suppor dari masing masing itemset
    support = {}
    num_transactions = float(len(dataset))
    for itemset in itemsets:
        count = 0
        for transaction in dataset:
            if itemset.issubset(transaction):
                count += 1
        support[itemset] = count / num_transactions
    # Filter itemset dengan support yang kurang dari min_support
    support = {itemset: support[itemset] for itemset in support if support[itemset] >= min_support}
    return support

def create_candidates(itemsets, k):
    # Buat sebuah list kandidat itemsets dengan ukuran k
    candidates = set()
    for itemset1 in itemsets:
        for itemset2 in itemsets:
            if len(itemset1.union(itemset2)) == k:
                candidates.add(itemset1.union(itemset2))
    return candidates

def apriori(dataset, min_support):
    # Generate frequent itemsets menggunkakan algoritma apriori
    itemsets = create_itemsets(dataset)
    support = calculate_support(dataset, itemsets, min_support)
    frequent_itemsets = [itemsets]
    k = 2
    while len(frequent_itemsets[k-2]) > 0:
        candidates = create_candidates(frequent_itemsets[k-2], k)
        itemsets = calculate_support(dataset, candidates, min_support)
        frequent_itemsets.append(itemsets)
        k += 1
    return frequent_itemsets

# Eksekusi Program
dataset = load_dataset()
frequent_itemsets = apriori(dataset, 0.4)
for k, itemsets in enumerate(frequent_itemsets):
    print("Frekuensi itemsets untuk size {}:{}".format(k+1, itemsets))
```
